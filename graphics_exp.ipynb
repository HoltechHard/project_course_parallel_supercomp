```python
# packages
import numpy as np
import pandas as pd
import math
import matplotlib.pyplot as plt
import seaborn as sns
```

Experiment 01

```python
# data

k = [i for i in range(1, 6)]

abs_err1 = [0.103881102062960, 0.000269169948388, 0.000119386415226, 
                0.000000247545428, 0.000000139331969]

rel_err1 = [5.194055103148010, 0.013458497419383, 0.005969320761290, 
                0.000012377271397, 0.000006966598454]

time1 = [6.90680800, 0.08389700, 0.07353600, 0.08926300, 0.10419600]                
```

```python
# in logaritm10 scale
sns.set()
plt.plot(k, np.log10(abs_err1))
plt.plot(k, np.log10(rel_err1))
plt.title("Experiment 01: Analysis k versus error (in logaritmic scale)")
plt.ylabel("Logaritm(10) of absolute error")
plt.xlabel("k: degree of quadrature")
plt.legend(title= "errors", loc = "upper right", labels = ["absolute", "relative"])
```

```python
plt.plot(k, np.log10(time1))
plt.title("Experiment 01: Analysis k versus time")
plt.ylabel("Time (in logaritm10 scale)")
plt.xlabel("k: degree of quadrature")
```
